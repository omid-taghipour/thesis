# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score
from matplotlib.colors import ListedColormap
```

# Preprocessing Data set

* Importing dataset

```python
cleveland_dataset = pd.read_csv('heart_2020_cleaned.csv')
```

* Seprating target column from Features

```python
X = cleveland_dataset.iloc[:, :1].values
y = cleveland_dataset.iloc[:, :1].values
```

```python
print(X)
```

* Splitting to Train and Test dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1, stratify=y)
```

* Feature Scaling

```python
sc = StandardScaler()
X_train[:,8:] = sc.fit_transform(X_train[:,8:])
X_test[:,8:] = sc.fit_transform(X_test[:,8:])
```

# Implementation of SVM Models

#### 1. Linear SVM

1. 1. Train the model

```python
classifier = SVC(kernel='linear', random_state=1)
classifier.fit(X_train, y_train)
```

1. 2. Prediction

```python
y_predict = classifier.predict(X_test)
```

1. 3. Accuracy of the model

```python
cm = confusion_matrix(y_test, y_predict)
print(cm)
accuracy = accuracy_score(y_test, y_predict)
# Converting accuracy to percentage
print(accuracy * 100)
```
